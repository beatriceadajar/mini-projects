# Postpaid Plans

I am planning on buying a new phone and switching to a postpaid plan once I've got the money and my affairs in order. I wanted to see the price differences for each plan over the course of four years (the longest I'm probably going to hold on to a phone, given past experiences, although I will probably want to switch to a new phone after 3 years).

I checked out a few options for getting the Huawei P20 (my target phone) and a postpaid plan for either Smart or Globe. I then calculated the total spending I would make over four years to see which one would be overall the best option given price and data/call/text provisions.

Here are my results.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
mpl.rcParams['figure.dpi'] = 80
```

## List of Possible Options

| Provider | Postpaid Plan                   | Description                                            | Price               |
|:--------:|:-------------------------------:|:------------------------------------------------------:|:-------------------:|
| Globe    | Huawei P20 + ThePLAN 599        | 2.5 GB Data, Unli All-Net Text                         | P18,599 + 599/month |

## Scenario 1

Assume that I won't go over the provided data/call/text allotment. This is what the cost would look like over the course of 4 years (each dot on the graph signifies the end of a year):

```python
# equations
# a: y = 599x + 18000
# b: y = 799x + 15600
# c: y = 399x + 34990
# d: y = 599x + 24000

mpl.style.use('seaborn')
plt.suptitle('Postpaid Plans Total Cost Over 4 Years')
plt.xlabel('Month')
plt.ylabel('Total Cost')

x = np.linspace(0, 48)
x2 = np.linspace(12, 48, 4)

plt.plot(x, 599*x + 18000, 'C1', label='ThePLAN 599')
plt.plot(x, 799*x + 15600, 'C2', label='ThePLAN 799')
plt.plot(x, 399*x + 34990, 'C3', label='Giga Plan Line 399')
plt.plot(x, 599*x + 24000, 'C4', label='GigaX Plan 599')

plt.plot(x2, 599*x2 + 18000, 'C1', marker='o')
plt.plot(x2, 799*x2 + 15600, 'C2', marker='o')
plt.plot(x2, 399*x2 + 34990, 'C3', marker='o')
plt.plot(x2, 599*x2 + 24000, 'C4', marker='o')

plt.legend()
plt.show()
```

Insights:

1. Overall, ThePLAN 599 is the cheapest option, both at the end of the 24-month contract and for the rest of the months after.
2. ThePLAN 799 and Giga Plan Line 399 cost about the same at the end of four years. So if I plan on keeping my phone for four years, and I find that ThePLAN 599 doesn't provide enough for me, I should just opt for ThePLAN 799 rather than buying the device itself and getting Giga Plan Line 399.

Based on Scenario 1, I think I'd go for **ThePLAN 599.**

## Scenario 2

In case you haven't noticed, ThePLAN 599 doesn't include calls in its plan. This doesn't bother me, as I almost never make calls on my phone, but it would also be naive to assume that, in four years, I would _never_ need to make a call. Thus, for this scenario, I will assume that I will make at most 20 minutes worth of calls per month, which is the allotment for the Smart plans, half of which are to a network that is not Globe/TM. This would increase ThePLAN 599 by P140/month, and ThePLAN 799 by P75/month (since it only provides unli-call for Globe/TM). This would be the output of the new equations:

```python
# equations
# a: y = 739x + 18000
# b: y = 874x + 15600
# c: y = 399x + 34990
# d: y = 599x + 24000

mpl.style.use('seaborn')
plt.suptitle('Postpaid Plans Total Cost Over 4 Years (With Additional Charges)')
plt.xlabel('Month')
plt.ylabel('Total Cost')

x = np.linspace(0, 48)
x2 = np.linspace(12, 48, 4)

plt.plot(x, 739*x + 18000, 'C1', label='ThePLAN 599')
plt.plot(x, 874*x + 15600, 'C2', label='ThePLAN 799')
plt.plot(x, 399*x + 34990, 'C3', label='Giga Plan Line 399')
plt.plot(x, 599*x + 24000, 'C4', label='GigaX Plan 599')

plt.plot(x2, 739*x2 + 18000, 'C1', marker='o')
plt.plot(x2, 874*x2 + 15600, 'C2', marker='o')
plt.plot(x2, 399*x2 + 34990, 'C3', marker='o')
plt.plot(x2, 599*x2 + 24000, 'C4', marker='o')

plt.legend()
plt.show()
```

Insights:

1. By the end of the 24-month contract, ThePLAN 599 is still the cheapest option, although the difference with ThePLAN 799 is now almost negligible.
2. By the end of four years, however, the GigaX Plan 599 becomes the cheapest option, although the difference with ThePLAN 599 is even more negligible.
3. ThePLAN 799 remains the most expensive option, but in this scenario, is no longer tied with any other plan.

Although the GigaX Plan 599 is the cheapest in four years, the amount of money I'd save with ThePLAN 599 if I discontinue my plan before four years outweighs the money I'd lose if I kept the plan for four years.

Thus, based on Scenario 2, I'd still pick **ThePLAN 599.**
